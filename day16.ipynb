<h2>Data Visualization :-</h2>

Tool :- Matplotlib --> python library / Module

Matplotlib is a Python library used for creating static, animated, and interactive visualizations in a wide range of formats.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("libraries imported")
```

```python
country_population = [700,750,740,730,752,800,790]
country_population
```

```python
# Line Plot
plt.plot(country_population)
plt.show()
```

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker='*',ms=18,mfc='black') # marker is used to represent points in plot , ms = marker size , mfc = marker face color
plt.show()
```

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.xlabel('Axis-x')
plt.ylabel('Axis-y')
plt.plot(x_axis,country_population,marker='P',ms=18,mfc='yellow') # marker is used to represent points in plot , ms = marker size , mfc = marker face color
plt.show()
```

<h3>BAR PLOT :-</h3>

used when one axis is catgorical , one axis is numerical

```python
country_population = [450,700,900,500,800]
country_name = ['Japan','Bangladesh','India','Nepal','China']
plt.bar(country_name,country_population)
plt.xlabel("Country_Name")
plt.ylabel("( in millions )")
plt.show()
```

```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['brand'].value_counts()
```

```python
brand_name = list(df['brand'].value_counts().keys())
brand_name
```

```python
no_of_count = list(df['brand'].value_counts().values)
print(no_of_count)
```

```python
brand_name = list(df['brand'].value_counts().keys())
no_of_count = list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count)
plt.xticks(rotation=90)
# plt.yticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(10,4)) # (width,height) to change the dimensions of the figure at output
brand_name = list(df['brand'].value_counts().keys())
no_of_count = list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count,color='red',ec='black') # ec = edge color
plt.xticks(rotation=90)
# plt.yticks(rotation=75)
plt.savefig('brand_population') # to save image with the specific name provided
plt.show()
```
