```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
# 1 --> Diabetic
# 0 --> Non-Diabetic
```

```python
x = df.drop(columns='Outcome',axis=1)
x
```

```python
y = df['Outcome']
y
```

```python
# Splitting Train Test Data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=23)
```

```python
print(x_test.shape,x_train.shape)
```

```python
# Model Training 
model_diabetic = LogisticRegression()
```

```python
# Training model with the training data

model_diabetic.fit(x_train,y_train)
```

```python
y_pred = model_diabetic.predict(x_test)
```

```python
score = accuracy_score(y_test,y_pred)
```

```python
print("The Accuracy : ",score*100)
```

```python
# Making a Predictive System

input = (1,85,66,29,0,26.6,0.351,31)

arr = np.array(input)
```

```python
print(arr)
```

```python
prediction = model_diabetic.predict(arr.reshape(1,-1))
```

```python
if prediction == 0:
    print("You are Healthy")

else:
    print("Talk to Doctor")
```
