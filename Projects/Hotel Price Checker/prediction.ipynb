```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("hotels.csv")
df.head()
```

```python
df.City.value_counts().keys()
```

```python
df.shape
```

```python
df[df['City']=='jaipur'].sort_values(by='Hotel_Rating',ascending=False)
```

```python
df2 = df[['Hotel_Rating','City','Hotel_Price']]
df2.head()
```

```python
df2.isna().sum()
```

```python
df2.duplicated().sum()
```

```python
le = LabelEncoder()
```

```python
df2['City_encoded'] = le.fit_transform(df2['City'])
```

```python
df2.head(20)
```

```python
df2.drop('City',axis=1,inplace=True)
```

```python
df2.head()
```

```python
df2.City_encoded.value_counts().keys()
```

```python
x = df2.drop('Hotel_Price',axis=1)
```

```python
y = df2[['Hotel_Price']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
lnr = LinearRegression()
rfr = RandomForestRegressor()
dtr = DecisionTreeRegressor()
```

```python
lnr.fit(x_train,y_train)
rfr.fit(x_train,y_train)
dtr.fit(x_train,y_train)
```

```python
print("Training Score : ",lnr.score(x_train,y_train)*100)
print("Testing Score : ",lnr.score(x_test,y_test)*100)
```

```python
print("Training Score : ",rfr.score(x_train,y_train)*100)
print("Testing Score : ",rfr.score(x_test,y_test)*100)
```

```python
print("Training Score : ",dtr.score(x_train,y_train)*100)
print("Testing Score : ",dtr.score(x_test,y_test)*100)
```
